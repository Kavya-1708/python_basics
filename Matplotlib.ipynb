Q1. Generate a line plot graph for the column 'model' and 'hp'

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv("/content/cars-1.csv")  # Replace 'your_file.csv' with your actual file name

# Step 2: Create the line plot for 'model' and 'hp'
plt.figure(figsize=(10, 6))
plt.plot(df['model'], df['hp'], marker='o')  # Line plot

# Step 3: Add labels and title
plt.xlabel('Model of cars')
plt.ylabel('Horse-Power of cars')
plt.title('Model Names vs HorsePower')
plt.xticks(rotation=45, ha = "right")


# Step 4: Display the plot
plt.show()
```

Q2. Create a bar plot graph for the columns 'carbs' and 'gear'

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv("/content/cars-2.csv")  # Replace 'your_file.csv' with your actual file name

# Step 2: Select data for the bar chart
# Assuming the CSV file has columns like 'Category' and 'Values'
carbs = df['carb']  # Replace with the appropriate column name
gear= df['gear']  # Replace with the appropriate column name

# Step 3: Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(carbs,gear, color='skyblue')

# Step 4: Add labels and title
plt.xlabel('Number of carburetors')
plt.ylabel('Number of forward gears')
plt.title('Carbs vs gear')

# Step 5: Display the chart
plt.xticks(rotation=45, ha="right")  # Rotate labels to prevent overlap
plt.tight_layout()  # Adjust layout to fit everything nicely
plt.show()
```

Q3. Plot a histogram for the column 'wt'

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv("/content/cars-3.csv")  # Replace 'your_file.csv' with the actual file name

# Step 2: Select the column you want to create a histogram for
data_column = df['wt']  # Replace 'YourColumnName' with the column you'd like to visualize

# Step 3: Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(data_column, bins=30, color='skyblue', edgecolor='black')  # Adjust bins as needed

# Step 4: Add labels and title
plt.xlabel('weight of the cars')
plt.ylabel('count')
plt.title('Histogram for the weight values')

# Step 5: Display the histogram
plt.tight_layout()  # Adjust layout to fit everything nicely
plt.show()
```

Q4. Plot a pie chart for columns: 'cyl' and 'model'

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv("/content/cars-4.csv")  # Replace with actual file name

# Step 2: Aggregate data
cyl_counts = df['cyl'].value_counts()  # Count occurrences in 'cyl' column
model_counts = df['model'].value_counts()  # Count occurrences in 'model' column

# Step 3: Create a figure with two pie charts
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for 'cyl' column
axes[0].pie(cyl_counts, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'coral', 'gold'])
axes[0].set_title("Cylinder Distribution")

# Pie chart for 'model' column
axes[1].pie(model_counts, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'orange', 'purple', 'red'])
axes[1].set_title("Model Distribution")

# Adjust layout for clarity
plt.tight_layout()

# Display the charts
plt.show()
```
