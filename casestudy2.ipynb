Q1.  bar-plot for the contract column.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the data
df = pd.read_csv("/content/customer_churn.csv")  # Replace 'your_file.csv' with the actual file name

# Step 2: Count occurrences in the 'Contract' column
contract_counts = df['Contract'].value_counts()  # Count the unique contract types

# Step 3: Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(contract_counts.index, contract_counts.values, color='orange')

# Step 4: Customize the plot
plt.xlabel('Contract Type of customer')  # X-axis label
plt.ylabel('Count')  # Y-axis label
plt.title('Distribution of Contract')  # Title

# Step 5: Display the plot
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout
plt.show()
```

Q2. histogram for the ‘MonthlyCharges’ column.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv("/content/customer_churn.csv")  # Replace 'your_file.csv' with the actual file name

# Step 2: Extract the 'MonthlyCharges' column
monthly_charges = df['MonthlyCharges']  # Replace 'MonthlyCharges' with the correct column name

# Step 3: Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(monthly_charges, bins=10, color='forestgreen', edgecolor='black')  # Assign 'forestgreen' color to bins

# Step 4: Customize the plot
plt.xlabel('Monthly Charges Incurred')  # X-axis label
plt.ylabel('Count')  # Y-axis label
plt.title('Distribution of Monthly Charges')  # Title

# Step 5: Display the histogram
plt.tight_layout()  # Adjust layout for clarity
plt.show()
```

Q3. scatter plot between TotalCharge

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('customer_churn.csv')

# Convert TotalCharges to numeric (if it's not already)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Drop rows with missing TotalCharges or tenure
df = df.dropna(subset=['TotalCharges', 'tenure'])

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['tenure'], df['TotalCharges'], color='indigo')

# Set axis labels and title
plt.xlabel('Tenure of the customer')
plt.ylabel('Total chargesIncurred')
plt.title('Total Charges vs Tenure')

# Show the plot
plt.grid(True)
plt.show()
```


4. Sam has to build a box-plot between ‘MonthlyCharges’ and ‘PaymentMethod’.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('customer_churn.csv')

# Drop rows with missing values in relevant columns
df = df.dropna(subset=['PaymentMethod', 'MonthlyCharges'])

# Create the box plot
plt.figure(figsize=(12, 6))
df.boxplot(column='MonthlyCharges', by='PaymentMethod', patch_artist=True,
           boxprops=dict(facecolor='olive'))

# Set axis labels and title
plt.xlabel('Payment Method of customer')
plt.ylabel('Monthly ChargesIncurred')
plt.title('Monthly Charges vs. Payment Method')

# Remove the automatic subplot title
plt.suptitle('')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.grid(True)
plt.show()
```
